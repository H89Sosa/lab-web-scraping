````python
#3. Use string manipulation techniques to replace whitespaces and linebreaks (i.e. \n) 
#   in the text of each html element. Use a list to store the clean names.

def dev_names(lst):
    '''
    This function takes a list of strings.
    Filters the ones which has two skipline at the end of the string, which we're interested
    For each string, extracts only the words.
    If the string has just one word , we format it with (@word) and appends it to 'account'
    If the string has two words, joins them and appends resulting string to 'name'
    Then, creates a zip with name and account and appends it propery to a list
    '''
    name = []
    account = []
    devs = []
    for word in lst:
        
        # our useful info has \n\n at the end of the string
        if re.search('\n{2}$', word):  
            word = word.split()
            
            if len(word) == 1: 
                # 1 string means account
                word = '(@' + word[0] + ')'
                account.append(word)
                
            else:     
                word = ' '.join(word)
                name.append(word)
                
    for name, account in zip(name, account):
        info = name + ' ' + account
        devs.append(info)
        
    return devs
    
dev_name_user = dev_names(text)
````

#### Retrieve an arbitary Wikipedia page of "Python" and create a list of links on that page

````python
# This is the url you will scrape in this exercise
url ='https://en.wikipedia.org/wiki/Python' 
````

````python
#your code
html = requests.get(url).content

soup = BeautifulSoup(html, 'lxml')
````

````python
tags = ['a']
````

````python
text = [element for element in soup.find_all(tags, href=True)]

text
````

````python
def clean_url(lst):
    '''
    This function finds links on a list
    '''
    links = []
    for string in lst:
        string= str(string)
        words = re.split('[\s | ; | > |]', string)
        for word in words:
            if 'http' in word:
                links.append(word[6:])
    return links
    
````

````python
clean_url(text)
````
